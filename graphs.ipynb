```python
import numpy as np
import matplotlib.pyplot as plt

#ok so plan of attack
#1. find avg by dept (write function to do this)
#2. find avg for course in the past
#3. find avg for prof in the past

N = 4
lows = (3,4,5,3)
avgs = (6,7,5,8)
highs = (10,11,12,11)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, lows, width, color='#d62728')
p2 = plt.bar(ind, avgs, width,
             bottom=lows, color = '#f442cb')
p3 = plt.bar(ind, avgs, width,
             bottom=avgs, color = '#63cbe8')

plt.ylabel('Amount of time spent')
plt.title('Comparison of time spent on this course to other related courses')
plt.xticks(ind, ('this course', 'this course over time', 'this dept', 'this professor'))
#plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0], p3[0]), ('Low', 'Average', 'High'))

plt.show()
```
